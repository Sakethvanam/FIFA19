```python
import pyspark 
from pyspark.sql import SparkSession 

  
#create instance of spark class 
spark=SparkSession.builder.appName('FIFA19').getOrCreate()
df=spark.read.csv("/FileStore/tables/Fifa19.csv",inferSchema=True,header=True)
df.show(20)
```

```python
df.printSchema()
```

```python
df.columns
```

```python
from pyspark.ml.linalg import Vectors 
from pyspark.ml.feature import VectorAssembler 
assembler=VectorAssembler(inputCols=['Potential','Stamina','Strength','Balance','Reactions'],outputCol='features')  
assembler.setHandleInvalid("skip").transform(df).show
output=assembler.transform(df) 
output.select('features','Overall').show(10) 
```

```python
final_data=output.select('features','overall')
```

```python
final_data.head()
```

```python
train_data ,test_data = final_data.randomSplit([0.80,0.20]) 
print(train_data.count())
print(test_data.count())
train_data.describe().show() 
```

```python
from pyspark.ml.regression import LinearRegression 
model=LinearRegression(featuresCol='features',labelCol='overall')
```

```python
trained_model=model.fit(train_data)  
overall_results=trained_model.evaluate(train_data)
```

```python
print('R-squared Error :',overall_results.r2)
```

```python
unlabeled_data=test_data.select('features')
unlabeled_data.show(5)
```

```python
predictions=trained_model.transform(unlabeled_data) 
predictions.show(50) #Predicting Most Promising Players
```

```python
from pyspark.mllib.evaluation import RegressionMetrics
print("MSE = %s" % overall_results.meanSquaredError)
```

```python
print("RMSE = %s" % overall_results.rootMeanSquaredError)
```

```python
print("MAE = %s" % overall_results.meanAbsoluteError)
print("Explained variance = %s" % overall_results.explainedVariance)
```
